```python
# Función para formatear números en millones o billones
def formatear_monto(valor):
    if abs(valor) >= 1_000_000_000:
        return '{:.2f}B'.format(valor / 1_000_000_000)
    elif abs(valor) >= 1_000_000:
        return '{:.2f}M'.format(valor / 1_000_000)
    else:
        return '{:,.2f}'.format(valor)

# Aplicar la función al campo 'Monto'
dtExport['Sales'] = dtExport['Sales'].apply(formatear_monto)
```

```python
dtExport['Tickets Sold'] = dtExport['Tickets Sold'].apply(formatear_monto)
```
