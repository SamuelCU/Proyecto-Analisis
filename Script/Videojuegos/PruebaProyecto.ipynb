```python
dataframe2 = pd.read_csv('games-data.csv')
dataframe2
```

```python
dataframe2['platform'] = dataframe2['platform'].astype(str).str.replace("Dreamcast", "DC")
dataframe2['platform'] = dataframe2['platform'].astype(str).str.replace("GameBoyAdvance", "GBA")
dataframe2['platform'] = dataframe2['platform'].astype(str).str.replace("GameCube", "GC")
dataframe2['platform'] = dataframe2['platform'].astype(str).str.replace("Nintendo64", "N64")
dataframe2['platform'] = dataframe2['platform'].astype(str).str.replace("PlayStation", "PS")
dataframe2['platform'] = dataframe2['platform'].astype(str).str.replace("PlayStation2", "PS2")
dataframe2['platform'] = dataframe2['platform'].astype(str).str.replace("PlayStation3", "PS3")
dataframe2['platform'] = dataframe2['platform'].astype(str).str.replace("PlayStation4", "PS4")
dataframe2['platform'] = dataframe2['platform'].astype(str).str.replace("PlayStationVita", "PSV")
dataframe2['platform'] = dataframe2['platform'].astype(str).str.replace("Xbox360", "X360")
dataframe2['platform'] = dataframe2['platform'].astype(str).str.replace("Xbox", "XB")
```

```python
dataframe2['r-date'] = pd.to_datetime(dataframe2['r-date'])

# Extraer el año y crear una nueva columna 'year'
dataframe2['year'] = dataframe2['r-date'].dt.year
```

```python
dataframe2.drop('r-date', axis=1,inplace=True)
```

```python
dataframe2['first_genre'] = dataframe2['genre'].str.split(',').str[0]
```

```python
dataframe2.drop('genre', axis=1,inplace=True)
```

```python
dataframe2['players'].fillna('No info', inplace=True)
```
