```python
import pandas as pd
```

```python
df1 = pd.read_csv("most popular sport in the world 2023.csv", header=0, delimiter=";")
df1
```

```python
df3 =pd.read_csv("The 30 Most Popular Sports in the World 2023.csv", header=0, delimiter=";", index_col='Rank')
df3
```

```python

df3['Number of fans'] = df3['Number of fans'].astype(str).str.replace(',', '')


df3['Number of fans']= pd.to_numeric(df3['Number of fans'], errors='coerce')
df3.dtypes
```

```python
valor1 =df3.at[19-1, 'Number of fans']
valor2=df3.at[19+1, 'Number of fans']

valorpromedio = (valor1+valor2)/2
df3.at[19, 'Number of fans'] = valorpromedio

df3
```

```python
valor3 =df3.at[21-1, 'Number of fans']
valor4=df3.at[21+2, 'Number of fans']

valorpromedio = (valor3+valor4)/2
df3.at[21, 'Number of fans'] = valorpromedio

df3 
```

```python
valor3 =df3.at[22-1, 'Number of fans']
valor4=df3.at[22+1, 'Number of fans']

valorpromedio = (valor3+valor4)/2
df3.at[22, 'Number of fans'] = valorpromedio

df3 
```

```python
valor3 =df3.at[27-1, 'Number of fans']
valor4=df3.at[27+2, 'Number of fans']

valorpromedio = (valor3+valor4)/2
df3.at[27, 'Number of fans'] = valorpromedio

df3 
```

```python
valor3 =df3.at[28-1, 'Number of fans']
valor4=df3.at[28+1, 'Number of fans']

valorpromedio = (valor3+valor4)/2
df3.at[28, 'Number of fans'] = valorpromedio

df3 
```
