```python
import pandas as pd
```

```python
df1 = pd.read_csv("most popular sport in the world 2023.csv", header=0, delimiter=";", index_col='Rank')
df1
```

```python
df2 =pd.read_csv("The 30 Most Popular Sports in the World 2023.csv", header=0, delimiter=";", index_col='Rank')
df2
```

```python
df2 = df2.rename(columns={'Number of fans': 'Estimated Fans'})
df2
```

```python
df_concatenado= pd.concat([df1, df2])
df_concatenado
```

```python
df3 =pd.read_csv("Sport by country.csv", header=0, delimiter=";")
df3
```

```python
df4 =pd.read_csv("watched on TV.csv", header=0, delimiter=";",index_col='Sport')
df4
```

```python
df4 = df4.drop('Unnamed: 2', axis=1)
df4
```

```python
df4 = df4.drop('Unnamed: 3', axis=1)
df4
```

```python
df_c = pd.merge(df_concatenado, df4, on='Sport', how='outer')
df_c
```

```python
df_c.to_json('df_final.json', orient='records')
```

```python
import couchdb
import pandas as pd
import json

# Conectar al servidor CouchDB
couch = couchdb.Server('http://estefania:Fhachi116@localhost:5984')

# Seleccionar la base de datos
db = couch['deportes']

# Leer todos los documentos (excluyendo metadatos)
rows = db.view('_all_docs', include_docs=True)

# Extraer solo los datos de los documentos
data = [row['doc'] for row in rows]

# Convertir los datos a un DataFrame de pandas
df = pd.DataFrame(data)

# Exportar el DataFrame a un archivo JSON
df.to_json('datos2.json', orient='records')
```
